```python
def f1(x):
    return 2*x*x-x-1

def f1_dao(x):
    return 4*x-1

def f2(x):
    return 3*x*x-21.6*x-1

def f2_dao(x):
    return 6*x-21.6

```

## 1. 使用黄金分割法求解第2题，alpha，beta为区间端点

```python

x1=golden_section(f1,[-1,1],0.06)
print('(a)黄金分割法返回的x为'+str(x1)+'，函数值为'+str(f1(x1)))
print('--------')
x2=golden_section(f2,[0,25],0.08)
print('(b)黄金分割法返回的x为'+str(x2)+'，函数值为'+str(f2(x2)))
```

## 2. 使用Fibonacci求解第2题，alpha，beta为区间端点

```python

x1=fibonacci_search(f1,[-1,1],0.06)
print('(a)Fibonacci返回的x为'+str(x1)+'，函数值为'+str(f1(x1)))
print('--------')
x2=fibonacci_search(f2,[0,25],0.08)
print('(b)Fibonacci返回的x为'+str(x2)+'，函数值为'+str(f2(x2)))
```

## 3. 使用二分法求解第2题，alpha，beta为区间端点，lambda为区间中点，dao_list为各中点的导数值

```python

x1=bin_split(f1_dao,[-1,1],0.06)
print('(a)二分法返回的x为'+str(x1)+'，函数值为'+str(f1(x1)))
print('--------')
x2=bin_split(f2_dao,[0,25],0.08)
print('(b)二分法返回的x为'+str(x2)+'，函数值为'+str(f2(x2)))
```

## 4. 使用dichotomous求解第2题，alpha，beta为区间端点，lambda为区间中点-eps，mu为区间中点+eps

```python
x1=dichotomous(f1,[-1,1],0.06,1e-4)
print('(a)dichotomous返回的x为'+str(x1)+'，函数值为'+str(f1(x1)))
print('--------')
x2=dichotomous(f2,[0,25],0.08,1e-4)
print('(b)dichotomous返回的x为'+str(x2)+'，函数值为'+str(f2(x2)))
```

结论：以上算法都可对单谷函数计算最小值，但与实际情况有一定的误差，此误差与用户给定的区间精度参数有关
